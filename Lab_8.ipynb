# Minimum Cost Spanning Tree
Problem Description:
A spanning tree of a graph can be defined as a graph with minimal set of edges that connect all vertices. A minimum spanning tree of a graph is a spanning tree of the graph with least weight (where the weight is computed by adding the weights of all the edges in the spanning tree). In general, a graph can have multiple minimum spanning trees. The problem is to find a minimum spanning tree of a graph.

This implementation uses a dictionary to represent the weighted graph, where the vertices are keys, and the values are dictionaries containing their adjacent vertices and their weights.

Please note that this implementation assumes that the input graph is connected and weighted.

First, let's create a simple graph to demonstrate Prim's algorithm:

```python
graph = {
    "A": {"B": 10, "C": 15, "D": 20},
    "B": {"A": 10, "C": 12},
    "C": {"A": 15, "B": 12, "D": 18},
    "D": {"A": 20, "C": 18},
}
```

Now, let's implement Prim's algorithm:

```python
def prim_algorithm(graph):
    if not graph:
        return {}, 0

    # Start with an arbitrary vertex (the first one in the graph)
    start_vertex = list(graph.keys())[0]
    visited = {start_vertex}
    min_tree_edges = [] # Stores edges of the MST as (u, v, weight)
    total_weight = 0

    num_vertices = len(graph)

    # An MST for a connected graph with V vertices has V-1 edges.
    # We continue until we have V-1 edges or cannot find more connections.
    while len(min_tree_edges) < num_vertices - 1:
        min_edge_weight = float('inf')
        best_edge = None # Stores (u, v, weight) for the best edge found in this iteration

        # Iterate through all vertices already in the MST (i.e., in 'visited')
        for u in visited:
            # Iterate through all neighbors of 'u'
            for v, weight in graph[u].items():
                # If 'v' is not yet visited and the edge (u, v) is the minimum found so far
                if v not in visited and weight < min_edge_weight:
                    min_edge_weight = weight
                    best_edge = (u, v, weight)

        # If no new edge was found, it implies the graph is not connected or all reachable nodes are visited.
        # For this problem, we assume the graph is connected, so this should not happen before V-1 edges are found.
        if best_edge is None:
            break

        # Add the selected edge to the MST
        u, v, weight = best_edge
        visited.add(v)
        min_tree_edges.append((u, v, weight))
        total_weight += weight

    # Convert the list of edges to an adjacency list dictionary format for the MST
    mst_dict = {vertex: [] for vertex in graph}
    for u, v, weight in min_tree_edges:
        mst_dict[u].append((v, weight))
        mst_dict[v].append((u, weight)) # MST edges are undirected, add both directions

    return mst_dict, total_weight
```

Now, let's run the Prim's algorithm on the example graph:
Keep in mind that this example uses a simple representation of a graph, but for more complex cases, you might want to use a library such as NetworkX to represent and manipulate graphs.

```python
mst, min_weight = prim_algorithm(graph)
print("Minimum Spanning Tree: ", mst)
print("Minimum Weight: ", min_weight)
```

```python
class DisjointSetUnion:
    def __init__(self, vertices):
        self.parent = {v: v for v in vertices}
        self.rank = {v: 0 for v in vertices}

    def find(self, i):
        if self.parent[i] == i:
            return i
        self.parent[i] = self.find(self.parent[i])
        return self.parent[i]

    def union(self, i, j):
        root_i = self.find(i)
        root_j = self.find(j)

        if root_i != root_j:
            if self.rank[root_i] < self.rank[root_j]:
                self.parent[root_i] = root_j
            elif self.rank[root_i] > self.rank[root_j]:
                self.parent[root_j] = root_i
            else:
                self.parent[root_j] = root_i
                self.rank[root_i] += 1
            return True
        return False

def kruskal_algorithm(graph):
    edges = []
    for u in graph:
        for v, weight in graph[u].items():
            # Add each edge once (e.g., only for u < v to avoid duplicates in undirected graph)
            if u < v:
                edges.append((weight, u, v))

    # Sort edges by weight
    edges.sort()

    vertices = list(graph.keys())
    dsu = DisjointSetUnion(vertices)
    mst_edges = []
    total_weight = 0
    num_vertices = len(vertices)

    for weight, u, v in edges:
        if len(mst_edges) == num_vertices - 1:
            break

        if dsu.union(u, v):
            mst_edges.append((u, v, weight))
            total_weight += weight

    # Convert the list of edges to an adjacency list dictionary format for the MST
    mst_dict = {vertex: [] for vertex in graph}
    for u, v, weight in mst_edges:
        mst_dict[u].append((v, weight))
        mst_dict[v].append((u, weight)) # MST edges are undirected, add both directions

    return mst_dict, total_weight

# Run Kruskal's algorithm on the new_graph
mst_kruskal, min_weight_kruskal = kruskal_algorithm(new_graph)
print("Minimum Spanning Tree (Kruskal's Algorithm): ", mst_kruskal)
print("Minimum Weight (Kruskal's Algorithm): ", min_weight_kruskal)
```
