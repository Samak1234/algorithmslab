```python
# Exercise 4
# to find the number of common elements between two lists, a and b
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# write a functin named as linear_search to perfrom linear search in 1D-array
def linear_search(arr,target):
  for i in range(len(arr)):
    if arr[i]==target:
      return i
  return -1
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
arr=[1,3,5,7,9]
target=5
print(linear_search(arr,target))
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def binary_search(arr,target):
  low=0
  high=len(arr)-1
  while low<=high:
    mid=(low+high)//2 # Changed / to // for integer division
    if(arr[mid]==target):
      return mid
    elif(arr[mid]>target):
     high=mid+1
    else:
      low=mid+1

  return -1
```
