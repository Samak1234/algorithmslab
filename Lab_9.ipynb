# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
# Modified dijkstra function to also return predecessors for path reconstruction
def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    predecessors = {node: None for node in graph} # To reconstruct the path
    distances[start] = 0
    unvisited = set(graph.keys()) # Set of unvisited nodes for efficient lookup

    while unvisited:
        # Find the unvisited node with the smallest distance
        min_node = None
        min_distance = float('inf')

        for node in unvisited:
            if distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None or min_distance == float('inf'):
            break # All remaining unvisited nodes are unreachable

        unvisited.remove(min_node)

        for neighbor, weight in graph[min_node].items():
            # Ensure neighbor is a known node in the graph (important if graph definition can be incomplete)
            if neighbor in graph:
                new_distance = distances[min_node] + weight
                if new_distance < distances[neighbor]:
                    distances[neighbor] = new_distance
                    predecessors[neighbor] = min_node # Update predecessor
    return distances, predecessors

# Helper function to reconstruct the shortest path
def get_shortest_path(predecessors, start, target):
    path = []
    current = target

    # Traverse backwards from target to start using predecessors
    while current is not None and current != start:
        path.insert(0, current) # Add current node to the beginning of the path
        current = predecessors[current]

    # If we successfully reached the start node, add it to the path
    if current == start:
        path.insert(0, start)
    else: # If current is None, it means the target is unreachable from start, or start was not in predecessors if start == target
        if start == target and start in predecessors: # Special case: if start and target are the same and start is a valid node
            return [start] # Path is just the start node itself
        return [] # Path not found
    return path

# Function to get user input for the graph
def get_user_graph():
    graph = {}
    print("Enter graph details. Nodes can be single letters (e.g., A, B, C).")
    print("Type 'done' for node name when you're finished adding nodes.")

    while True:
        node_name_input = input("Enter node name: ").strip().upper()
        if node_name_input == 'DONE':
            break
        if not node_name_input.isalpha() or len(node_name_input) != 1:
            print("Invalid node name. Please enter a single letter.")
            continue

        node_name = node_name_input
        if node_name not in graph: # Ensure node is added only once
            graph[node_name] = {}

        print(f"Enter neighbors and weights for node {node_name}. Format: 'Neighbor Weight' (e.g., B 5).")
        print("Type 'done' when you're finished adding neighbors for this node.")

        while True:
            neighbor_input = input(f"  Neighbor and weight for {node_name}: ").strip().upper()
            if neighbor_input == 'DONE':
                break

            parts = neighbor_input.split()
            if len(parts) == 2:
                neighbor, weight_str = parts[0], parts[1]
                if not neighbor.isalpha() or len(neighbor) != 1:
                    print("  Invalid neighbor name. Please enter a single letter.")
                    continue
                try:
                    weight = int(weight_str)
                    if weight < 0:
                        print("  Weight cannot be negative for Dijkstra's algorithm. Please enter a non-negative integer.")
                        continue
                    graph[node_name][neighbor] = weight
                except ValueError:
                    print("  Invalid weight. Please enter an integer.")
            else:
                print("  Invalid input format. Please use 'Neighbor Weight' (e.g., B 5).")
    return graph

# --- Demonstration and Testing ---
print("--- Dijkstra Algorithm Implementation ---")

# Option for user input graph
user_choice = input("Do you want to enter your own graph (yes/no)? ").strip().lower()

my_graph = {}
if user_choice == 'yes':
    my_graph = get_user_graph()
    if not my_graph:
        print("No graph entered by user. Using a predefined sample graph instead.")
        user_choice = 'no' # Fallback to predefined graph
else:
    print("Using a predefined sample graph.")

if user_choice == 'no':
    # Predefined sample graph with added nodes E and F
    my_graph = {
        'A': {'B': 4, 'C': 2},
        'B': {'A': 4, 'C': 5, 'D': 10, 'E': 2}, # Added E
        'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8}, # Added F
        'D': {'B': 10, 'C': 3, 'E': 6},
        'E': {'B': 2, 'D': 6, 'F': 1}, # New Node E
        'F': {'C': 8, 'E': 1}  # New Node F
    }
    print("\nPredefined graph:")
    for node, neighbors in my_graph.items():
        print(f"  {node}: {neighbors}")

# Test algorithm with different source nodes
print("\n--- Testing Dijkstra's Algorithm ---")

# Define source nodes to test
source_nodes_to_test = ['A', 'B', 'E'] # Include new node E

for source_node in source_nodes_to_test:
    if source_node in my_graph:
        distances, predecessors = dijkstra(my_graph, source_node)
        print(f"\nShortest paths from source node '{source_node}':")
        for target_node in my_graph:
            path = get_shortest_path(predecessors, source_node, target_node)
            distance = distances[target_node]

            if distance == float('inf'):
                print(f"  To {target_node}: Unreachable")
            else:
                print(f"  To {target_node}: Distance = {distance}, Path = {' -> '.join(path)}")
    else:
        print(f"Source node '{source_node}' not found in the graph.")
```
